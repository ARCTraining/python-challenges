## Probability distribution



```python
import random
import seaborn as sns
sns.set_theme()
```

### Uniform Distribution

```python
# INTEGER
distribution = [random.randint(1, 100) for _ in range(1000000)]
sns.histplot(data=distribution, bins=100)
```

```python
# REAL
distribution = [random.uniform(1, 100) for _ in range(1000000)]
sns.histplot(data=distribution)
```

```python
# SEQUENCE
distribution = random.choices(['A','B', 'C', 'D'], k=100000)
sns.histplot(data=distribution, shrink=.8)
```

### Weighted

```python
distribution = random.choices(['A','B', 'C', 'D'], weights=[1,1,1,2], k=100000)
sns.histplot(data=distribution, shrink=.8)
```

### Gaussian

```python
distribution = [random.gauss(0, 2) for _ in range(1000000)]
sns.histplot(data=distribution)
```

### Others

```python
distribution = [random.triangular() for _ in range(1000000)]
sns.histplot(data=distribution)
```

```python
distribution = [random.betavariate(1,2) for _ in range(1000000)]
sns.histplot(data=distribution)
```

```python
distribution = [random.expovariate(1) for _ in range(1000000)]
sns.histplot(data=distribution)
```
